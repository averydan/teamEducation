```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
poverty_rate_2019_df = pd.read_csv('Resources/poverty19.csv')
poverty_rate_2018_df = pd.read_csv('Resources/poverty18.csv')
poverty_rate_2017_df = pd.read_csv('Resources/poverty17.csv')
poverty_rate_2016_df = pd.read_csv('Resources/poverty16.csv')
poverty_rate_2015_df = pd.read_csv('Resources/poverty15.csv')

poverty_rate_2015_df.head()
```

```python
# school_dist_19 = poverty_rate_2019_df.groupby('Name')
poverty_rate_2019_df['Poverty Rate'] = round(poverty_rate_2019_df['Estimated number of relevant children 5 to 17 years old in poverty who are related to the householder']/poverty_rate_2019_df['Estimated Population 5-17'] * 100, 2)
poverty_rate_2018_df['Poverty Rate'] = round(poverty_rate_2018_df['Estimated number of relevant children 5 to 17 years old in poverty who are related to the householder']/poverty_rate_2018_df['Estimated Population 5-17'] * 100, 2)
poverty_rate_2017_df['Poverty Rate'] = round(poverty_rate_2017_df['Estimated number of relevant children 5 to 17 years old in poverty who are related to the householder']/poverty_rate_2017_df['Estimated Population 5-17'] * 100, 2)
poverty_rate_2016_df['Poverty Rate'] = round(poverty_rate_2016_df['Estimated number of relevant children 5 to 17 years old in poverty who are related to the householder']/poverty_rate_2016_df['Estimated Population 5-17'] * 100, 2)
poverty_rate_2015_df['Poverty Rate'] = round(poverty_rate_2015_df['Estimated number of relevant children 5 to 17 years old in poverty who are related to the householder']/poverty_rate_2015_df['Estimated Population 5-17'] * 100, 2)

columns_to_drop = ['State Postal Code',
                  'State FIPS Code',
                  'District ID',
                  'Estimated Total Population',
                  'Estimated Population 5-17',
                  'Estimated number of relevant children 5 to 17 years old in poverty who are related to the householder']

poverty_rate_2019_df.drop(columns_to_drop, axis=1, inplace=True)
poverty_rate_2018_df.drop(columns_to_drop, axis=1, inplace=True)
poverty_rate_2017_df.drop(columns_to_drop, axis=1, inplace=True)
poverty_rate_2016_df.drop(columns_to_drop, axis=1, inplace=True)
poverty_rate_2015_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
combined_poverty_rate = pd.merge(poverty_rate_2019_df, poverty_rate_2018_df,
                                 how='outer', on='Name')
combined_poverty_rate = combined_poverty_rate.rename(columns={
    'Poverty Rate_x' : '2019 Poverty Rate',
    'Poverty Rate_y' : '2018 Poverty Rate'
})

combined_poverty_rate = pd.merge(combined_poverty_rate, poverty_rate_2017_df,
                                 how='outer', on='Name')

combined_poverty_rate = combined_poverty_rate.rename(columns={
    'Poverty Rate' : '2017 Poverty Rate'
})
combined_poverty_rate = pd.merge(combined_poverty_rate, poverty_rate_2016_df,
                                 how='outer', on='Name')

combined_poverty_rate = combined_poverty_rate.rename(columns={
    'Poverty Rate' : '2016 Poverty Rate'
})
combined_poverty_rate = pd.merge(combined_poverty_rate, poverty_rate_2015_df,
                                 how='outer', on='Name')

combined_poverty_rate = combined_poverty_rate.rename(columns={
    'Poverty Rate' : '2015 Poverty Rate'
})
combined_poverty_rate
```
