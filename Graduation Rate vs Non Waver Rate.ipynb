```python
plt.figure(figsize=(15,5))
barWidth = 0.3
 
# set height of bar
bars1 = grad_rates2015['State Non-Waiver Grad Rate']
bars2 = grad_rates2015['State Grad Rate']

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
names = grad_rates2015['Corporation Name']
 
# Make the plot
plt.bar(r1, bars1, color='indianred', width=barWidth, edgecolor='white', label='State Non-Waiver Grad Rate')
plt.bar(r2, bars2, color='lightseagreen', width=barWidth, edgecolor='white', label='State Grad Rate')
 
# Add xticks on the middle of the group bars
plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))],names, rotation=45)
plt.xlabel('School Corporations')  
plt.ylabel('Graduation Percentages')



# Create legend & Show graphic
plt.legend(loc='lower right')
# plt.show()
```

```python
plt.figure(figsize=(20,22))
barWidth = 0.3
 
# set height of bar
bars1 = grad_rates2015['State Non-Waiver Grad Rate']
bars2 = grad_rates2015['State Grad Rate']

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x - barWidth for x in r1]

 
# Make the plot
plt.bar(r1, bars1, color='indianred', width=barWidth, edgecolor='white', label='State Non-Waiver Grad Rate', zorder=3)
plt.bar(r2, bars2, color='lightseagreen', width=barWidth, edgecolor='white', label='State Grad Rate', zorder=3)
 
# Add xticks on the middle of the group bars
# plt.xlabel('group', fontweight='bold')
plt.xticks(np.arange(0,len(bars1)),names, rotation = 90, fontsize=5, ha= 'center')
plt.xlabel('School Corporations')  
plt.ylabel('Graduation Percentages')
plt.grid(zorder=0)
plt.ylim(45,102)
plt.title('2014 Graduation Rates vs. Non-Waiver Graduation Rates')


# Create legend & Show graphic
plt.legend(loc='lower right')

plt.savefig('Output_Data/2014 Grad vs Non-Waiver Grad Rates.png')
plt.show()
```
