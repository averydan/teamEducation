# Graduation Rate vs Non Wavier Rate

```python
import pandas as pd
```

```python
grad_rates2014 = pd.read_csv('Resources/2014_grad_rates.csv')
grad_rates2015 = pd.read_csv('Resources/2015_grad_rates.csv')
grad_rates2016 = pd.read_csv('Resources/2016_grad_rates.csv')
grad_rates2017 = pd.read_csv('Resources/2017_grad_rates.csv')
grad_rates2018 = pd.read_csv('Resources/2018_grad_rates.csv')
grad_rates2019 = pd.read_csv('Resources/2019_grad_rates.csv')
grad_rates2019.head()
```

```python
grad_rates2014 = grad_rates2014.rename(columns={'Non-Waiver\nGrad N' : 'Non-Waiver Grad N' ,
                                                 'Non-Waiver\nGrad Rate': 'Non-Waiver Grad Rate'})
grad_rates2014.head()
```

```python
grad_rates2015 = grad_rates2015.rename(columns={'Non-Waiver\nGrad N' : 'Non-Waiver Grad N' ,
                                                 'Non-Waiver\nGrad Rate': 'Non-Waiver Grad Rate'})
grad_rates2015.head()
```

```python
grad_rates2016 = grad_rates2016.rename(columns={'Non-Waiver\nGrad N' : 'Non-Waiver Grad N' ,
                                                 'Non-Waiver\nGrad Rate': 'Non-Waiver Grad Rate'})
grad_rates2016.head()
```

```python
grad_rates2018 = grad_rates2018.rename(columns={'Corp Id' : 'Corp ID',
                                                'CORPORATION NAME' : 'Corporation Name',
                                                '2018 State Grad Rate' : 'Grad Rate',
                                                 '2018 State Non-Waiver Grad Rate': 'Non-Waiver Grad Rate'})
grad_rates2018.head()
```

```python
grad_rates2019 = grad_rates2019.rename(columns={'Corp Id' : 'Corp ID',
                                                'CORPORATION NAME' : 'Corporation Name',
                                                '2019 State Grad Rate' : 'Grad Rate',
                                                 '2019 State Non-Waiver Grad Rate': 'Non-Waiver Grad Rate'})
grad_rates2019.head()
```
