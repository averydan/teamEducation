```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
grad_rates2014 = pd.read_csv('Resources/2014_grad_rates.csv')
grad_rates2015 = pd.read_csv('Resources/2015_grad_rates.csv')
grad_rates2016 = pd.read_csv('Resources/2016_grad_rates.csv')
grad_rates2017 = pd.read_csv('Resources/2017_grad_rates.csv')
grad_rates2018 = pd.read_csv('Resources/2018_grad_rates.csv')
grad_rates2019 = pd.read_csv('Resources/2019_grad_rates.csv')
grad_rates2019.head()
```

```python
grad_rates2014 = grad_rates2014.rename(columns={'Corp Name' : 'Corporation Name',
                                                'Non-Waiver\nGrad N' : 'Non-Waiver Grad N' ,
                                                 'Non-Waiver\nGrad Rate': 'Non-Waiver Grad Rate'})
grad_rates2014.head()
```

```python
grad_rates2015 = grad_rates2015.rename(columns={'Non-Waiver\nGrad N' : 'Non-Waiver Grad N' ,
                                                 'Non-Waiver\nGrad Rate': 'Non-Waiver Grad Rate'})
grad_rates2015.head()
```

```python
grad_rates2016 = grad_rates2016.rename(columns={'Non-Waiver\nGrad N' : 'Non-Waiver Grad N' ,
                                                 'Non-Waiver\nGrad Rate': 'Non-Waiver Grad Rate'})
grad_rates2016.head()
```

```python
grad_rates2018 = grad_rates2018.rename(columns={'Corp Id' : 'Corp ID',
                                                'CORPORATION NAME' : 'Corporation Name',
                                                '2018 State Grad Rate' : 'Grad Rate',
                                                 '2018 State Non-Waiver Grad Rate': 'Non-Waiver Grad Rate'})
grad_rates2018.head()
```

```python
grad_rates2019 = grad_rates2019.rename(columns={'Corp Id' : 'Corp ID',
                                                'CORPORATION NAME' : 'Corporation Name',
                                                '2019 State Grad Rate' : 'Grad Rate',
                                                 '2019 State Non-Waiver Grad Rate': 'Non-Waiver Grad Rate'})
grad_rates2019.head()
```

```python
columns_to_drop = ['Corp ID',
                  'Cohort N',
                  'Grad N',
                  'Non-Waiver Grad N']
grad_rates2014.drop(columns_to_drop, axis=1, inplace=True)
grad_rates2015.drop(columns_to_drop, axis=1, inplace=True)
grad_rates2016.drop(columns_to_drop, axis=1, inplace=True)
grad_rates2017.drop(columns_to_drop, axis=1, inplace=True)
grad_rates2018.drop(columns_to_drop, axis=1, inplace=True)
grad_rates2019.drop(columns_to_drop, axis=1, inplace=True)
```

```python
combined_grad_rate = pd.merge(grad_rates2014, grad_rates2015, how = 'outer', on = 'Corporation Name')

combined_grad_rate = combined_grad_rate.rename(columns={
    'Grad Rate_x' : '2014 Grad Rate',
    'Non-Waiver Grad Rate_x' : '2014 Non-Waiver Rate',
    'Grad Rate_y' : '2015 Grad Rate',
    'Non-Waiver Grad Rate_y' : '2015 Non-Waiver Rate'})

combined_grad_rate = pd.merge(combined_grad_rate, grad_rates2016, how = 'outer', on = 'Corporation Name')

combined_grad_rate = combined_grad_rate.rename(columns={
    'Grad Rate' : '2016 Grad Rate',
    'Non-Waiver Grad Rate' : '2016 Non-Waiver Rate'})

combined_grad_rate = pd.merge(combined_grad_rate, grad_rates2017, how = 'outer', on = 'Corporation Name')

combined_grad_rate = combined_grad_rate.rename(columns={
    'Grad Rate' : '2017 Grad Rate',
    'Non-Waiver Grad Rate' : '2017 Non-Waiver Rate'})

combined_grad_rate = pd.merge(combined_grad_rate, grad_rates2018, how = 'outer', on = 'Corporation Name')

combined_grad_rate = combined_grad_rate.rename(columns={
    'Grad Rate' : '2018 Grad Rate',
    'Non-Waiver Grad Rate' : '2018 Non-Waiver Rate'})

combined_grad_rate = pd.merge(combined_grad_rate, grad_rates2019, how = 'outer', on = 'Corporation Name')

combined_grad_rate = combined_grad_rate.rename(columns={
    'Grad Rate' : '2019 Grad Rate',
    'Non-Waiver Grad Rate' : '2019 Non-Waiver Rate'})

column_school_drop = ['School Name']
combined_grad_rate.drop(column_school_drop, axis=1, inplace=True)

combined_grad_rate.head()
```
